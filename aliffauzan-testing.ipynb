```python
import json
import requests
import base64
import tensorflow as tf

# 1. Siapkan data dummy (sample pasien diabetes)
# Catatan: Karena model kita menerima raw TF Examples (tf.string), 
# kita harus me-serialize dan men-encode data ini ke format Base64.
sample_data = {
    "Pregnancies": tf.train.Feature(int64_list=tf.train.Int64List(value=[6])),
    "Glucose": tf.train.Feature(int64_list=tf.train.Int64List(value=[148])),
    "BloodPressure": tf.train.Feature(int64_list=tf.train.Int64List(value=[72])),
    "SkinThickness": tf.train.Feature(int64_list=tf.train.Int64List(value=[35])),
    "Insulin": tf.train.Feature(int64_list=tf.train.Int64List(value=[0])),
    "BMI": tf.train.Feature(float_list=tf.train.FloatList(value=[33.6])),
    "DiabetesPedigreeFunction": tf.train.Feature(float_list=tf.train.FloatList(value=[0.627])),
    "Age": tf.train.Feature(int64_list=tf.train.Int64List(value=[50]))
}

# 2. Bungkus ke dalam tf.train.Example
example = tf.train.Example(features=tf.train.Features(feature=sample_data))

# 3. Serialize lalu encode ke Base64 (Syarat mutlak TF Serving)
serialized_example = example.SerializeToString()
encoded_example = base64.b64encode(serialized_example).decode('utf-8')

# 4. Siapkan struktur payload JSON
payload = {
    "signature_name": "serving_default",
    "instances": [
        {
            "b64": encoded_example
        }
    ]
}

# 5. Tembak URL Railway lu (jangan lupa pakai endpoint :predict)
url = "https://aliffauzan-mlops-submission-production.up.railway.app/v1/models/diabetes-model:predict"

print(f"Mengirim request ke: {url}...\n")

# 6. Kirim POST request
response = requests.post(url, json=payload)

# 7. Tampilkan hasil prediksi
print("Status Code:", response.status_code)
print("Hasil Prediksi:", response.json())
```
